```python
import numpy as np

# --- 1. Fungsi Inti ---
def bind(v1, v2):
    """Operasi Binding (Enkripsi Cermin)"""
    return v1 * v2

def compute_sim(v1, v2):
    """Menghitung kemiripan antara dua vektor"""
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

# --- 2. Inisialisasi Data & Kunci ---
D = 10000
data_asli = np.random.choice([-1, 1], size=D)
kunci_koordinat = np.random.choice([-1, 1], size=D)

# --- 3. Proses Mirroring (Enkripsi) ---
# Menciptakan 'Pantulan' yang akan disimpan di sistem
pantulan = bind(data_asli, kunci_koordinat)

# --- 4. Verifikasi Keamanan ---
sim_asli_pantulan = compute_sim(data_asli, pantulan)

print("=== ANALISIS KEAMANAN MIRROR ===")
print(f"Similarity (Data Asli vs Pantulan): {sim_asli_pantulan:.6f}")
print("Hasil: " + ("AMAN (Data tidak terbaca)" if abs(sim_asli_pantulan) < 0.02 else "RESIKO (Data bocor)"))

# --- 5. Proses Rekonstruksi (Dekripsi) ---
# Menggunakan kunci yang sama untuk memulihkan data
data_terpulih = bind(pantulan, kunci_koordinat)

# Cek keutuhan data (Integrity Check)
is_perfect = np.array_equal(data_asli, data_terpulih)
print("\n=== ANALISIS INTEGRITAS DATA ===")
print(f"Apakah data berhasil dipulihkan 100%? {is_perfect}")
```
